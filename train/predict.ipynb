```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

def show_result(image, limit = 0.5):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32)
    image /= 255.0
    image = torch.tensor(image).permute(2,0,1)
    image = image.unsqueeze(dim=0)
    # image = image.reshape(1,3,1376,1024)
    image = image.float().to(device)
    # print(image.shape)

    model.eval()
    output = model(image)
    # print(output)

    img = image.squeeze().permute(1,2,0).cpu()
    img = np.array(img)
    # print(img.shape)

    for idx, b in enumerate(output[0]['boxes']):
        box = b.cpu().detach().numpy()
        x1 = box[0]
        y1 = box[1]
        x2 = box[2]
        y2 = box[3]
        lbl = output[0]['labels'][idx].cpu().detach().numpy()
        scr = output[0]['scores'][idx].cpu().detach().numpy()
        # print(x1, y1, x2, y2, lbl, scr)
        if scr > limit:
            img = cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0,0,255), 1)
            img = cv2.putText(img, str(lbl) , (int(x1), int(y1)), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,0,255), 1)

    plt.figure(figsize=(10,20))
    plt.imshow(img)
    
image = plt.imread("./books/chudo_derevo_redmi/IMG_20190715_112936.labeled.jpg")
# image = plt.imread("./books/chudo_derevo_redmi/IMG_20190715_112912.labeled.jpg")
print(type(image))
show_result(image, limit=0.5)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

def draw_img(file_txt, file_img, filter=None):
    image = plt.imread(file_img)
    print(image.shape)
    (height, width, ch) = image.shape
    # (width, height, ch) = image.shape

    with open(file_txt, 'r') as f:
        ls = f.readlines()
        filted_lbl_lst = []
        for la in ls:
            (lbl, pcx, pcy, pw, ph) = la.strip().split(' ')
            
            font_color = (0,0,255)
            if filter != None and int(lbl) in filter:
                filted_lbl_lst.append(int(lbl))
                font_color = (255, 0,0)
            
            cx = width * float(pcx)
            cy = height * float(pcy)
            w = width * float(pw)
            h = height * float(ph)

            x1 = cx - w/2
            y1 = cy - h/2
            x2 = cx + w/2
            y2 = cy + h/2
            
            # print(x1, y1, x2, y2)
            
            img = cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), font_color, 1)        
            img = cv2.putText(img, str(lbl) , (int(x1), int(y1)), cv2.FONT_HERSHEY_SIMPLEX, 0.3, font_color, 1)
            

    if filter == None or len(filted_lbl_lst) > 0:
        print(file_txt, filted_lbl_lst)
        plt.figure(figsize=(10,20))
        plt.imshow(image)
        plt.show()
        
file_txt = '/opt/ml/datasets/coco128/labels/train2017/000000000025.txt'
file_img = '/opt/ml/datasets/coco128/images/train2017/000000000025.jpg'
# file_txt = '/opt/ml/datasets/coco128/labels/train2017/000000000009.txt'
# file_img = '/opt/ml/datasets/coco128/images/train2017/000000000009.jpg'

draw_img(file_txt, file_img)
```
